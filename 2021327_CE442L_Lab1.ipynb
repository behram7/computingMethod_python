```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function and limits
a = -np.pi
b = np.pi
n = 11
delta = (b - a) / (n - 1)

# Generate points for the function and for the Riemann sum
x = np.linspace(a, b, 1000)  # Smooth curve points
x_riemann = np.linspace(a, b, n)  # Points for the Riemann sum
midpoints = (x_riemann[:-1] + x_riemann[1:]) / 2  # Midpoints for Riemann sum
y_mid = np.cos(midpoints)  # Cosine values at midpoints

# Calculate the Riemann sum using the midpoint rule
reimann = delta * sum(y_mid)

# Calculate the actual value of the integral
actualValue, _ = quad(np.cos, a, b)

# Print results
print("Midpoint Rule Approximation:", reimann)
print("Actual Value:", actualValue)
print("Difference:", abs(reimann - actualValue))
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, np.cos(x), label='y = cos(x)', color='blue', linewidth=2)  # Curve of cos(x)
plt.bar(midpoints, y_mid, width=delta, align='center', alpha=0.5, color='orange', label='Midpoint Riemann Sum')
plt.title('Midpoint Rule Riemann Sum Approximation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.xlim(a, b)
plt.ylim(-1.5, 1.5)

# Show the plot
plt.show()
```
