```python
import numpy as np
import sympy as sp

# Define the variable and function
x = sp.symbols('x')
func = sp.cos(x)

# Parameters
x0 = 0  # Center of the Taylor expansion
orders = [0, 2, 4, 6]  # Orders of the Taylor series

# Generate x values for plotting
x_values = np.linspace(-1, 7, 100)  # Range from -1 to 7

# Initialize a list to store the Taylor series values
taylor_values_list = []

# Loop through each specified order and calculate the Taylor series
for n in orders:
    # Initialize result for the current order
    result = 0
    for i in range(n + 1):
        deriv = func.diff(x, i).subs(x, x0)  # i-th derivative at x0
        result += (deriv / sp.factorial(i)) * ((x - x0) ** i)
    
    # Convert the Taylor series to a numerical function
    taylor_series_func = sp.lambdify(x, result, 'numpy')
    taylor_values = taylor_series_func(x_values)

    # For order 0, create a constant array
    if n == 0:
        taylor_values = np.ones_like(x_values)  # Create an array of ones
    
    # Append the values to the list
    taylor_values_list.append(taylor_values)

# Convert the list to a NumPy array
taylor_values_array = np.vstack(taylor_values_list)

# Print the resulting array
print("Taylor series values for orders 0, 2, 4, and 6:")
print(taylor_values_array)
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Define the variable and function
x = sp.symbols('x')
func = sp.cos(x)

x0 = 0  
orders = [0, 2, 4, 6]  

x_values = np.linspace(-1, 7, 100) 

taylor_values_list = []

for n in orders:
    # Initialize result for the current order
    result = 0
    for i in range(n + 1):
        deriv = func.diff(x, i).subs(x, x0)  # i-th derivative at x0
        result += (deriv / sp.factorial(i)) * ((x - x0) ** i)
    
    taylor_series_func = sp.lambdify(x, result, 'numpy')
    taylor_values = taylor_series_func(x_values)

    if n == 0:
        taylor_values = np.ones_like(x_values) 
    
    taylor_values_list.append(taylor_values)

taylor_values_array = np.vstack(taylor_values_list)

taylor_values_transposed = taylor_values_array.T

actual_values = np.cos(x_values)

x_target = 1.5
actual_value_at_target = np.cos(x_target)
taylor_values_at_target = [taylor_series_func(x_target) for taylor_series_func in [sp.lambdify(x, result, 'numpy') for n in orders]]

# Calculate errors
errors = [abs(actual_value_at_target - taylor_value) for taylor_value in taylor_values_at_target]

print(f"Actual value at x={x_target}: {actual_value_at_target}")
for i, order in enumerate(orders):
    print(f"Taylor Series (Order {order}) value: {taylor_values_at_target[i]}, Error: {errors[i]}")

plt.figure(figsize=(10, 6))
for i, order in enumerate(orders):
    plt.plot(x_values, taylor_values_transposed[:, i], label=f'Taylor Series (Order {order})')

plt.plot(x_values, actual_values, label='cos(x)', color='red', linewidth=2)
plt.title('Taylor Series Approximation of cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.xlim(-1, 7)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
import numpy as np

def f(x):
    return x**3 - x - 1

# Bisection method implementation
def bisection_method(func, a, b, tol=1e-5, max_iter=100):
    if func(a) * func(b) >= 0:
        raise ValueError("Function must have different signs at a and b")
    
    for i in range(max_iter):
        c = (a + b) / 2 
        if func(c) == 0 or (b - a) / 2 < tol:
            return c  
            b = c  
        else:
            a = c  
    
    return (a + b) / 2  
a = 1
b = 2

root = bisection_method(f, a, b)
print(f"Root found: {root}")
```
